# Predictive Road Maintenance Using Machine Learning

## Phase 1a: Data Collection and Validation

### Objective
The goal of this phase is to collect, organize, and validate all datasets required for predicting future road maintenance needs. The datasets include historical NYC 311 road complaints, traffic volume, weather observations, and OpenStreetMap road network data. Each dataset is verified before preprocessing begins.

## Datasets Collected

1. NYC 311 Service Requests (2012–2019)
2. NYC 311 Service Requests (2020–2025)
3. NYC Historical Traffic Volume (2012–2025)
4. Historical Weather Data (2012–2025)


```python
import pandas as pd
```

### Load the 311 service requests (2012–2019)


```python
# Load the 311 service requests (2012–2019)

service_311_2012_2019 = pd.read_csv(
    "../data/raw/311/311_service_requests_2012_2019.csv",
  
)
```

```python
# Display the first 5 rows

service_311_2012_2019.head()
```

```python
# Check the dataset

service_311_2012_2019.info()
```

```python
# Check the size

service_311_2012_2019.shape
```

```python
# Display the column names

service_311_2012_2019.columns
```

```python
# Count missing values

service_311_2012_2019.isnull().sum()
```

```python
# Check for duplicate rows

service_311_2012_2019.duplicated().sum()
```

```python
# Display all complaint types

service_311_2012_2019[
    "Problem (formerly Complaint Type)"
].value_counts()
```

```python
# Standardize the complaint type

service_311_2012_2019[
    "Problem (formerly Complaint Type)"
] = service_311_2012_2019[
    "Problem (formerly Complaint Type)"
].replace(
    "DEP Street Condition",
    "Street Condition"
)
```

```python
# Verify the complaint types

service_311_2012_2019[
    "Problem (formerly Complaint Type)"
].value_counts()
```

```python
# Check the earliest and latest complaint dates just to double check the data is within the expected range

service_311_2012_2019["Created Date"].min()
```

```python
service_311_2012_2019["Created Date"].max()
```
