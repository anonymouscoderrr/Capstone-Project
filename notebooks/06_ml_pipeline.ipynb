# Phase 1E: Machine Learning Pipeline and Model Preparation

## Objective

Validate the saved machine learning pipeline and confirm that it can be loaded and used for future predictions.

This phase prepares the selected model for Phase 2 API development.

```python

import pandas as pd
import joblib
```

```python
# Load final machine learning dataset

model_data = pd.read_csv(
    "../data/processed/model_data.csv"
)
```

```python
# Load saved Decision Tree pipeline

pipeline = joblib.load(
    "../models/decision_tree_model.pkl"
)
```

```python
# Display loaded pipeline

pipeline
```

```python
# Prepare features and target

X = model_data.drop(
    columns=[
        "Maintenance Risk",
        "Complaint Count"
    ]
)

y = model_data["Maintenance Risk"]
```

```python
# Create sample input records

sample_input = X.head(5)

sample_input
```

```python
# Generate sample predictions

sample_predictions = pipeline.predict(
    sample_input
)

sample_predictions
```

```python
# Compare sample predictions with actual values

prediction_check = pd.DataFrame({
    "Actual": y.head(5).values,
    "Predicted": sample_predictions
})

prediction_check
```

```python
# Save feature list used by the model

feature_columns = list(X.columns)

joblib.dump(
    feature_columns,
    "../models/feature_columns.pkl"
)

feature_columns
```

```python
# Verify saved feature list

loaded_features = joblib.load(
    "../models/feature_columns.pkl"
)

loaded_features
```

```python
# Confirm pipeline is ready

print("Pipeline loaded successfully.")
print("Sample predictions completed successfully.")
print("Feature columns saved successfully.")
print("Model is ready for Phase 2 API development.")
```

## Phase 1E Summary

The final Decision Tree pipeline was loaded successfully and tested on sample input records. The feature column order was also saved for future use. The machine learning pipeline is now ready to be connected to a prediction API in Phase 2.
