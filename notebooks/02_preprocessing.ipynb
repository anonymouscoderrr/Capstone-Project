## Phase 1b: Data Standardization

In this notebook, the raw datasets will be cleaned and standardized so they can be used for feature engineering and machine learning.

The main goals are:
- clean the NYC 311 road complaint datasets
- standardize date columns
- standardize complaint types and road names
- remove duplicates
- handle missing values
- save cleaned datasets into the processed data folder

```python
import pandas as pd 
```

```python
# Load the 311 service requests dataset 


service_311_2012_2019 = pd.read_csv(
    "../data/raw/311/311_service_requests_2012_2019.csv", 
)

service_311_2020_2025 = pd.read_csv(
    "../data/raw/311/311_service_requests_2020_2025.csv",
)
```

```python
# Check the size of both datasets 

print(service_311_2012_2019.shape)
print(service_311_2020_2025.shape)
```

```python
# Combine the two 311 datasets

service_311 = pd.concat(
    [
        service_311_2012_2019,
        service_311_2020_2025
    ],
    ignore_index=True
)
```

```python
# Check the size of the combined dataset

service_311.shape
```

```python
# Display the first 5 rows

service_311.head()
```

```python
# Check for duplicates after we combined the dataset 
service_311.duplicated().sum()
```

```python
# Check missing values 

service_311.isnull().sum()
```

```python
# Display all column names

service_311.columns
```

```python
# Keep only the columns needed for the project

service_311 = service_311[
    [
        "Unique Key",
        "Created Date",
        "Closed Date",
        "Problem (formerly Complaint Type)",
        "Problem Detail (formerly Descriptor)",
        "Incident Zip",
        "Incident Address",
        "Street Name",
        "Cross Street 1",
        "Cross Street 2",
        "Address Type",
        "City",
        "Status",
        "Resolution Description",
        "Community Board",
        "Council District",
        "Police Precinct",
        "Borough",
        "Latitude",
        "Longitude"
    ]
]
```

```python
# Verify the remaining columns

service_311.columns
```

```python
# Check the new size

service_311.shape
```

```python
# Convert date columns to datetime

service_311["Created Date"] = pd.to_datetime(
    service_311["Created Date"],
    format="%m/%d/%Y %I:%M:%S %p",
    errors="coerce"
)

service_311["Closed Date"] = pd.to_datetime(
    service_311["Closed Date"],
    format="%m/%d/%Y %I:%M:%S %p",
    errors="coerce"
)
```

```python
# Check the data types

service_311.dtypes
```

```python
# Standardize text columns

text_columns = [
    "Problem (formerly Complaint Type)",
    "Problem Detail (formerly Descriptor)",
    "Incident Address",
    "Street Name",
    "Cross Street 1",
    "Cross Street 2",
    "Address Type",
    "City",
    "Status",
    "Resolution Description",
    "Community Board",
    "Police Precinct",
    "Borough"
]

for col in text_columns:
    service_311[col] = (
        service_311[col]
        .astype(str)
        .str.strip()
        .str.upper()
    )
```

```python
# Verify 
service_311[
    [
        "Street Name",
        "City",
        "Borough",
        "Status"
    ]
].head()
```

```python
# Check missing values in the cleaned 311 dataset

service_311.isnull().sum()
```

```python
# Check the Incident Zip data type

service_311["Incident Zip"].dtype
```

```python
# Convert Incident Zip to numeric

service_311["Incident Zip"] = pd.to_numeric(
    service_311["Incident Zip"],
    errors="coerce"
)
```

```python
# Verify the data type

service_311["Incident Zip"].dtype
```

```python
# Create Year from Created Date
service_311["Created Date"] = pd.to_datetime(service_311["Created Date"])

service_311["Year"] = service_311["Created Date"].dt.year
```

```python
# Create Month from Created Date
service_311["Month"] = service_311["Created Date"].dt.month
```

```python
# Keep only records through 2025

service_311 = service_311[
    service_311["Year"] <= 2025
]
```

```python
# Save the cleaned dataset

service_311.to_csv(
    "../data/processed/road_complaints.csv",
    index=False
)
```

```python
# Convert weather time to datetime
weather_raw["time"] = pd.to_datetime(
    weather_raw["time"],
    errors="coerce"
)

# Remove rows with invalid dates
weather_raw = weather_raw.dropna(subset=["time"])

# Create a daily date column
weather_raw["Date"] = weather_raw["time"].dt.normalize()

# Aggregate hourly weather into daily weather
weather = (
    weather_raw.groupby("Date", as_index=False)
    .agg(
        Avg_Temperature=("temperature_2m (°F)", "mean"),
        Total_Precipitation=("precipitation (inch)", "sum"),
        Total_Snowfall=("snowfall (inch)", "sum"),
        Avg_Snow_Depth=("snow_depth (ft)", "mean"),
        Avg_Wind_Speed=("wind_speed_10m (mp/h)", "mean")
    )
)

weather.head()
```

```python
print("Weather shape:", weather.shape)
print(weather.isnull().sum())

weather.head()
```

```python
traffic = (
    traffic_raw.groupby(
        ["Street Name", "Date"],
        as_index=False
    )
    .agg(
        Daily_Traffic_Volume=(
            "Daily_Traffic_Volume",
            "mean"
        ),
        Traffic_Records=(
            "SegmentID",
            "count"
        )
    )
)

traffic.head()
```

```python
weather.to_csv(
    "../data/processed/weather.csv",
    index=False
)

traffic.to_csv(
    "../data/processed/traffic.csv",
    index=False
)

print("Saved processed weather:", weather.shape)
print("Saved processed traffic:", traffic.shape)
```
